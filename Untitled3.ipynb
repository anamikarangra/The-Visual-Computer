```python






import os
import numpy as np
from matplotlib import pyplot as plt
import random

def load_img(img_dir, img_list):
    images = []
    for i, image_name in enumerate(img_list):
        if image_name.endswith('.npy'):
            image = np.load(os.path.join(img_dir, image_name))
            images.append(image)
    images = np.array(images)
    return images

def imageLoader(img_dir, img_list, mask_dir, mask_list, batch_size):
    L = len(img_list)
    while True:
        batch_start = 0
        batch_end = batch_size
        while batch_start < L:
            limit = min(batch_end, L)
            X = load_img(img_dir, img_list[batch_start:limit])
            Y = load_img(mask_dir, mask_list[batch_start:limit])
            yield (X, Y)
            batch_start += batch_size
            batch_end += batch_size
# Test the generator
train_img_dir = '/content/BraTS2020_TrainingData/input_data_128/train/images'
train_mask_dir = '/content/BraTS2020_TrainingData/input_data_128/train/masks'
train_img_list = sorted(os.listdir(train_img_dir))
train_mask_list = sorted(os.listdir(train_mask_dir))

batch_size = 2
train_img_datagen = imageLoader(train_img_dir, train_img_list, train_mask_dir, train_mask_list, batch_size)

# Verify generator.... In Python 3 next() is renamed as __next__()
img, msk = next(train_img_datagen)

print(f"Image Shape: {img.shape}")
print(f"Mask Shape: {msk.shape}")
print(f"Mask Max Value: {np.max(msk)}")
print(f"Mask Unique Values: {np.unique(msk)}")

img_num = random.randint(0, img.shape[0] - 1)
test_img = img[img_num]
test_mask = msk[img_num]
test_mask = np.argmax(test_mask, axis=3)

print(f"Test Mask Shape After Argmax: {test_mask.shape}")
print(f"Test Mask Max Value After Argmax: {np.max(test_mask)}")
print(f"Test Mask Unique Values After Argmax: {np.unique(test_mask)}")


n_slice = random.randint(0, test_mask.shape[2] - 1)
plt.figure(figsize=(12, 8))

plt.subplot(221)
plt.imshow(test_img[:, :, n_slice, 0], cmap='gray')
plt.title('Image flair')
plt.subplot(222)
plt.imshow(test_img[:, :, n_slice, 1], cmap='gray')
plt.title('Image t1ce')
plt.subplot(223)
plt.imshow(test_img[:, :, n_slice, 2], cmap='gray')
plt.title('Image t2')
plt.subplot(224)
plt.imshow(test_mask[:, :, n_slice])
plt.title('Mask')
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt
from keras.models import load_model
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import os # Import the os module to use os.path.exists

# Load model for predictions
my_model = load_model('/content/drive/MyDrive/newbrats_3d.hdf5', compile=False)

# Define paths and parameters
# Verify these paths to ensure they are correct. Use the 'ls' command in a
# code block below to list the contents of the directory and check for the presence of 'image_0.npy'
val_images_path = '/content/BraTS2020_TrainingData/input_data_128/val/images/'
val_masks_path = '/content/BraTS2020_TrainingData/input_data_128/val/masks/'
num_images = 100  # Assuming there are 100 validation images, adjust as needed

# Initialize lists to store metrics
accuracies = []
precisions = []
recalls = []
f1_scores = []

# Loop through all validation images
for img_num in range(num_images):
    test_img_path = val_images_path + 'image_' + str(img_num) + ".npy"
    test_mask_path = val_masks_path + 'mask_' + str(img_num) + ".npy"

    # Check if the files exist before attempting to load them
    if not os.path.exists(test_img_path):
        print(f"Warning: Image file not found: {test_img_path}")
        continue  # Skip to the next image if the current one is not found
    if not os.path.exists(test_mask_path):
        print(f"Warning: Mask file not found: {test_mask_path}")
        continue

    # Load image and mask
    test_img = np.load(test_img_path)
    test_mask = np.load(test_mask_path)
    test_mask_argmax = np.argmax(test_mask, axis=3)

    # Prepare input
    test_img_input = np.expand_dims(test_img, axis=0)

    # Predict
    test_prediction = my_model.predict(test_img_input)
    test_prediction_argmax = np.argmax(test_prediction, axis=4)[0, :, :, :]

    # Calculate metrics
    accuracy = accuracy_score(test_mask_argmax.flatten(), test_prediction_argmax.flatten())
    precision = precision_score(test_mask_argmax.flatten(), test_prediction_argmax.flatten(), average='weighted')
    recall = recall_score(test_mask_argmax.flatten(), test_prediction_argmax.flatten(), average='weighted')
    f1 = f1_score(test_mask_argmax.flatten(), test_prediction_argmax.flatten(), average='weighted')

    # Store metrics
    accuracies.append(accuracy)
    precisions.append(precision)
    recalls.append(recall)
    f1_scores.append(f1)

    # Plot individual slices from test predictions for verification (optional)
    n_slice = 55
    if n_slice < test_img.shape[2]:
        plt.figure(figsize=(12, 8))
        plt.subplot(231)
        plt.title('Testing Image')
        plt.imshow(test_img[:, :, n_slice, 1], cmap='gray')
        plt.subplot(232)
        plt.title('Testing Label')
        plt.imshow(test_mask_argmax[:, :, n_slice])
        plt.subplot(233)
        plt.title('Prediction on test image')
        plt.imshow(test_prediction_argmax[:, :, n_slice]) # Correct indentation, and use n_slice for consistency
        plt.show() # Add plt.show() to display the plot

# Calculate average metrics
avg_accuracy = np.mean(accuracies)
avg_precision = np.mean(precisions)
avg_recall = np.mean(recalls)
avg_f1 = np.mean(f1_scores)

# Print or use these averages as needed
print(f'Average Accuracy: {avg_accuracy}')
print(f'Average Precision: {avg_precision}')
print(f'Average Recall: {avg_recall}')
print(f'Average F1-score: {avg_f1}')
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from keras.models import load_model

# Load model for predictions
my_model = load_model('/content/drive/MyDrive/newbrats_3d.hdf5', compile=False)

# Function to load images and masks
def load_data(img_folder, mask_folder):
    images = []
    masks = []
    img_filenames = sorted(os.listdir(img_folder))
    mask_filenames = sorted(os.listdir(mask_folder))

    for img_filename, mask_filename in zip(img_filenames, mask_filenames):
        img_path = os.path.join(img_folder, img_filename)
        mask_path = os.path.join(mask_folder, mask_filename)

        img = np.load(img_path)
        mask = np.load(mask_path)

        images.append(img)
        masks.append(mask)

    return np.array(images), np.array(masks)

# Load the test dataset
test_img_folder = '/content/BraTS2020_TrainingData/input_data_128/val/images'
test_mask_folder = '/content/BraTS2020_TrainingData/input_data_128/val/masks'

test_images, test_masks = load_data(test_img_folder, test_mask_folder)
test_masks_argmax = np.argmax(test_masks, axis=4)

# Process and predict in batches
batch_size = 5
num_batches = len(test_images) // batch_size + int(len(test_images) % batch_size != 0)

all_predictions = []
for i in range(num_batches):
    start_idx = i * batch_size
    end_idx = min((i + 1) * batch_size, len(test_images))

    batch_images = test_images[start_idx:end_idx]
    batch_predictions = my_model.predict(batch_images)
    batch_predictions_argmax = np.argmax(batch_predictions, axis=4)

    all_predictions.extend(batch_predictions_argmax)

all_predictions = np.array(all_predictions)

# Plot a few slices from test predictions for verification
n_slice = 55
num_visualizations = min(5, len(test_images))  # Number of images to visualize

for i in range(num_visualizations):
    if n_slice >= test_images[i].shape[2]:
        raise ValueError(f"n_slice should be less than {test_images[i].shape[2]}")

    plt.figure(figsize=(12, 8))
    plt.subplot(231)
    plt.title(f'Testing Image {i}')
    plt.imshow(test_images[i][:, :, n_slice, 1], cmap='gray')
    plt.subplot(232)
    plt.title(f'Testing Label {i}')
    plt.imshow(test_masks_argmax[i][:, :, n_slice])
    plt.subplot(233)
    plt.title(f'Prediction on test image {i}')
    plt.imshow(all_predictions[i][:, :, n_slice])
    plt.show()


```
